```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

merged_dataset = pd.read_csv("../data/pop_gni_merged.csv", index_col='country')
merged_dataset.head()
```

```python
years = list(range(1800, 2051))
correlations = []

for year in years:
    pop_column = f"{year}_pop"
    gni_column = f"{year}_gni"
    correlation = merged_dataset[pop_column].corr(merged_dataset[gni_column])
    correlations.append(correlation)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(years, correlations, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Correlation')
plt.title('Correlation between Population and GNI over the years')
plt.grid(True)
plt.tight_layout()
plt.show()
```
