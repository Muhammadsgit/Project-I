## Which Country have the highest and lowest economic growth over the course of the dataset?

```python
growth_data = []


for index, row in df_gni.iterrows():
   
  
    for year in range(1800, 2040, 10):
        start_value = row[str(year)]
        end_value = row[str(year + 10)]
        
        growth_rate = ((end_value - start_value) / start_value) * 100
        growth_data.append([index, str(year), str(year + 10), growth_rate])

df_growth_rate = pd.DataFrame(growth_data, columns=['country', 'start_year', 'end_year', 'growth_rate'])
df_growth_rate = df_growth_rate.set_index('country')

df_growth_rate[(df_growth_rate['start_year'] == '2000') & (df_growth_rate['end_year'] == '2010')]
```

### Highest Economic Grown Rate from 2000 to 2010

```python
twothousands_to_twothousands_10 = df_growth_rate[(df_growth_rate['start_year'] == '2000') & (df_growth_rate['end_year'] == '2010')]
twothousands_to_twothousands_10 = twothousands_to_twothousands_10.sort_values(by='growth_rate', ascending=False)
plt.figure(figsize=(15,10))
sns.barplot(x=twothousands_to_twothousands_10.head(10).index.values, y=twothousands_to_twothousands_10.head(10)['growth_rate'])
plt.title('top 10 highest growth rate from 2000 to 2010')

#matplotlib.pyplot.text(x, y, s, fontdict=None, other key words)

plt.text(8, 800, "This plot showcases the countries with the highest GNI growth rates between 2000 and 2010.", ha='right', fontsize=10, color='blue')

plt.show()
```

**We can observe that Equatorial Ginea had 9x growtn in GNI.**

### Lowest Economic Growth Rate from 2000 to 2010

```python
twothousands_to_twothousands_10 = df_growth_rate[(df_growth_rate['start_year'] == '2000') & (df_growth_rate['end_year'] == '2010')]
twothousands_to_twothousands_10 = twothousands_to_twothousands_10.sort_values(by='growth_rate', ascending=True)
plt.figure(figsize=(15,10))
sns.barplot(x=twothousands_to_twothousands_10.head(10).index.values, y=twothousands_to_twothousands_10.head(10)['growth_rate'])
plt.title('top 10 highest growth rate from 2000 to 2010')
plt.text(8, -10, "This plot showcases the countries with the lowest GNI growth rates between 2000 and 2010.", ha='right', fontsize=10, color='blue')

plt.show()
```

We can observe that UAE had 15x decrease in GNI from 2000 tp 2010.

```python
# Creating fuction to convert billions, millions and thousands notations to intgers 
def convert_m_k(x) :
    if x[-1] == "M":
        x = x[:-1]
        x = float(x)* (10**6)


    elif x[-1] == "k":
        x = x[:-1]
        x = float(x)* (10**3)       
    elif x[-1] == "B":
        x = x[:-1]
        x = float(x)* (10**9)      
    return int(x)


df_check_for_MK = df_p2.applymap(lambda x: convert_m_k(x) )

df_check_for_MK
```

# Merging Both The Dataset

```python
df_check_for_MK.shape # Checking shapes for both databaser
```

```python
# inorder to merger, I am splicing the df two be of same colums

population_df_trimmed = df_check_for_MK.loc[:, :'2050'] 
population_df = population_df_trimmed.add_suffix('_pop')
gni_df = df_gni.add_suffix('_gni')

merged_df = population_df.merge(gni_df, left_index=True, right_index=True, how='inner')

ordered_columns=[]
for year in range(1800, 2051):
    ordered_columns.append(f"{year}_pop")
    ordered_columns.append(f"{year}_gni")

merged_df = merged_df[ordered_columns]
merged_df.to_csv('pop_gni_merged.csv')
merged_df

```
