Converting to integers.

```python
# Creating fuction to convert billions, millions and thousands notations to intgers 
def convert_m_k(x) :
    if x[-1] == "M":
        x = x[:-1]
        x = float(x)* (10**6)


    elif x[-1] == "k":
        x = x[:-1]
        x = float(x)* (10**3)       
    elif x[-1] == "B":
        x = x[:-1]
        x = float(x)* (10**9)   
    else : 
        x= float(x)    
    return int(x)


df_check_for_MK = df_p2.applymap(lambda x: convert_m_k(x) )

df_check_for_MK
```
