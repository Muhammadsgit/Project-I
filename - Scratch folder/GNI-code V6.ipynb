```python
def convert_m_k(x) :

    if type(x) == str and x[-1] == "k":
        x = x[:-1]
        x = float(x)* (10**3)       
    
    return int(x)


df_check_for_MK = df_gni.applymap(lambda x: convert_m_k(x) )

df_check_for_MK
```

Which Country have the highest and lowest economic growth over the course of the dataset?

```python
growth_data = []

# Loop through each country
for index, row in df_check_for_MK.iterrows():
    
    
    # Loop through the years in increments of 10 (from 1800 to 2040 to avoid extrapolated data)
    for year in range(1800, 2040, 10):
        start_value = row[str(year)]
        end_value = row[str(year + 10)]
        
        growth_rate = ((end_value - start_value) / start_value) * 100
        growth_data.append([index, str(year), str(year + 10), growth_rate])

# Convert the list to a DataFrame
df_growth_rate = pd.DataFrame(growth_data, columns=['country', 'start_year', 'end_year', 'growth_rate'])
df_growth_rate = df_growth_rate.set_index('country')

eighteen_hundred_to_eighteen_10 = df_growth_rate[(df_growth_rate['start_year'] == '1990') & (df_growth_rate['end_year'] == '2000')]
plt.figure(figsize=(45, 20))
sns.barplot(x=eighteen_hundred_to_eighteen_10.index.values[-25:], y = eighteen_hundred_to_eighteen_10['growth_rate'][-25:])
plt.show()
#df_growth_rate
```

```python
eighteen_hundred_to_eighteen_10_top_highest = eighteen_hundred_to_eighteen_10.sort_values(by='growth_rate', ascending=False)
plt.figure(figsize=(10,10))
sns.barplot(x=eighteen_hundred_to_eighteen_10_top_highest.head().index.values, y=eighteen_hundred_to_eighteen_10_top_highest.head()['growth_rate'])
plt.title('top 10 highest growth rate from 1990 to 2000')
plt.show()
eighteen_hundred_to_eighteen_10_top_highest.head()
```

```python
# Plotting the GNI growth trends for the top 10 countries in our dataset over the years using a scatter plot with all years on the x-axis

plt.figure(figsize=(17, 6))

for label, row in df_check_for_MK.iloc[:10,:].iterrows():
    plt.scatter(row.index, row.values, label=label, s=10)  # s=10 to set marker size

plt.title('GNI Growth Trends of Top 10 Countries in our dataset (1800-2050)')
plt.xlabel('Year')
plt.ylabel('GNI Growth')
plt.xticks(rotation=90, fontsize=5)
plt.legend()
plt.tight_layout()
plt.show()
```

### **Bar Plots**:

For a specific year, plot the GNI of each country.

## **Box Plots**:

Box plot for population growth rates of countries across decades.

```python

df_growth_rate_pivot = df_growth_rate.pivot(columns='start_year', values='growth_rate')
# The groth rate calculated above and converted to a data frame needs to be pivited for the box plot

# Plot box plots for population growth rates across decades
plt.figure(figsize=(20, 10))
sns.boxplot(data=df_growth_rate_pivot)
plt.title('Box Plot for GNI Growth Rates Across Decades')
plt.ylabel('Growth Rate')
plt.xlabel('Decade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
